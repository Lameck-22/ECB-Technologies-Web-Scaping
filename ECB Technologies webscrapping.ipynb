```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
base_url = 'https://ecbtechkenya.com/product-category/laptops/'
response = requests.get(base_url)
soup = BeautifulSoup(response.text,'html.parser')
```

```python
# Get the last element from the pagination list
page_links = soup.find( class_ = 'page-numbers').text.strip()
page_links = page_links.replace("Next Page", "")

max_page = int(page_links[-1])
print(f"Scraping {max_page} pages...")
```

```python
laptop_list = []
for page_num in range(1, max_page + 1):
    url = f"{base_url}page/{page_num}/"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    laptop_details = soup.find_all('div', class_ = 'mf-product-details')
    for laptop_detail in laptop_details:
        try:
            Name = laptop_detail.find('h2', class_ ='woo-loop-product__title').text.strip()
        except:
            Name = " "
        try:
            Brand = laptop_detail.find('div', class_='meta-brand').text.strip()
        except:
            Brand = " "
        try:
            Current_price = laptop_detail.find('span', class_='woocommerce-Price-amount amount').text.strip()
        except:
            Current_price = " "
        

        laptop_dict = {'Name':Name,
                    'Brand': Brand,
                    'Current_price': Current_price,}
        laptop_list.append(laptop_dict)

laptop_df = pd.DataFrame(laptop_list)
laptop_df
```

```python
#checking for duplicates
laptop_df.duplicated().sum()
```

```python
laptop_df.info()
```

```python
laptop_df['Current_price'] = laptop_df['Current_price'].replace('[KSh,]', '', regex=True).astype(float)
```

```python
laptop_df.info()
```

```python
# Remove the "Brand:" prefix from the Brand column
laptop_df['Brand'] = laptop_df['Brand'].str.replace('Brand:', '', regex=False).str.strip()

laptop_df
```

```python
laptop_df.to_csv('ECB laptops.csv', index = False)
```
